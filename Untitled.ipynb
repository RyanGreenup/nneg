# Non Linear Classification

## Load the Iris Data

Take the linear regression example from before

```r
N_dist = Distributions.Normal(0, 1)
N_dist_vec = Distributions.Normal(13, 100)
```

```r
using Gadfly, RDatasets

iris = dataset("datasets", "iris")
df = deepcopy(iris)

# Output Class
Y = Array{Int}(undef, size(df)[1], 2)
Y[:,1] = (df.Species .== "setosa") .+ 0
Y[:,2] = (df.Species .!= "setosa") .+ 0
Y[Y .== 0] .= -1

# Input Features
X = df[!, ["SepalLength", "SepalWidth"]]
X.bias .= 1
X = Matrix(X)

W = Random.rand(N_dist_vec, 2, 3)
```

## Plot the Data

```r
p2 = Gadfly.plot(df, x=:SepalLength, y=:SepalWidth, color=:Species, Geom.point)
```

```r
If we wanted to classify this type of data, an *SVM* would be perfect, however
```

```r
η = 1E-5
```

```r
for _ in 1:10000
    nrowdf = size(df)[1]
    ∇W_r = [0 0 0 ; 0 0 0]
    ε_r = 0
    for i in 1:nrowdf
        y = Y[i,:]
        x = X[i,:]

        yhat = W*Vector(x)
        ε_r += sum((y-yhat) .^ 2)
        ∇W_r += [-2*(y-yhat).*x[1:2]  -2*(y-yhat).*x[1:2]  -2*(y-yhat).*x[1:2]] / nrowdf
    end
    ε = ε_r
    W -= η*∇W_r
    # println(ε)
end
```
