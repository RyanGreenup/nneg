```python
from sympy import *
```

```python
y = [1]
@interact(x=(0,100))
def double_number(n):
    # Must be array as they are pass by reference
    # Well not really, it's called something else in Py.
    x = symbols('x')
    y[0] = (x+1)**n
    
    print(y)
```
